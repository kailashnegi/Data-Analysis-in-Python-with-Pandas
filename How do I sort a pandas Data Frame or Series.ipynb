```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.title.sort_values()
```

You can also sort a series by passing column name in brackets. See below.

```python
movies['title'].sort_values().head()
```

Sorting the data in descending order.

```python
movies['title'].sort_values(ascending=False)
```

##### Sorting the data frame with a particular column

```python
movies.sort_values('title').head(10)
```

You can sort Data Frame with a particular column in descending order as follows.

```python
movies.sort_values('star_rating',ascending=False).head(10)
```

You can also sort Data Frame with multiple columns. In this first Data Frame will be sorted based on content_rating and then based on genre. So for APPROVED content_ratings genre column will be sorted first den again sorted for different rating in content_rating column.

```python
movies.sort_values(['content_rating','genre'])
```
