```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.shape
```

###### Filtering rows based on a particular column value. Based on duration column in this case

```python
movies[movies['duration']>200]
```

You can also filter rows based on particular column value and then check it for Series as well. See below

```python
movies[movies['duration']>200].genre
```

It can also be done in following way.

```python
movies[movies['duration']>200]['genre']
```

It can also be done using loc which is very powerful method of doing it.

```python
movies.loc[movies.duration>200,'genre']
```

### Applying filter based on multiple columns

For example, if you want to filter movies where duration>=200 and genre is drama; it should be done as follows.

```python
movies[(movies.duration>=200) & (movies.genre=='Drama')]
```

What if we've to apply multiple OR conditions. The simple approach people will follow will be to apply multiple OR conditions i.e. (movies.genre=='Drama') OR (movies.genre=='Crime') OR (movies.genre=='Action') which is very lengthy. However there is a better approcah to implement it using "isin" function. See below.

```python
movies[movies.genre.isin(['Drama','Crime','Action'])].head(10)
```
