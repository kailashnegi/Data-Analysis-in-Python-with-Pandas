```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
#Use drop() method for droping columns or rows
ufo.drop('Colors Reported',axis=1,inplace=True)
```

```python
ufo.head()
```

If we've to drop multiple columns, we can do so by specifying all column names in a list and pass it to 
drop method.

```python
ufo.drop(['City','State'],axis=1,inplace=True)
```

```python
ufo.head()
```

If we've to drop rows, method is same. The only change is that instead of column names we have to pass row names and in place of axis=1, we've to pass axis=0. Although axis=0 is by default but passing it explicitly is recommended for better code understanding.

```python
#Removing rows
ufo.drop([0,1],axis=0,inplace=True)
```

```python
ufo.head()
```

We can also verify it by checking shape of the data frame. You'll see that shape of data frame has reduced.

```python
ufo.shape
```
